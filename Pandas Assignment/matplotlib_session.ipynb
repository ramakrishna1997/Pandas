```python
number = np.arange(0,5, 0.2)
```

```python
number
```

```python
plt.plot(number, number**2, 'b^', label = '^2')
plt.plot(number, number**2.4, 'r--', label = "^2.4")
plt.plot(number, number**3, 'gs', label = "^3")
plt.grid()
plt.legend()
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0, 5, 0.1)
t2 = np.arange(0, 0.5, 0.02)

plt.figure(1)

plt.subplot(121)
plt.grid()
plt.plot(t1, f(t1), 'r-')

plt.subplot(122)
plt.grid()
plt.plot(t2, f(t2), 'b^')

plt.show()
```
